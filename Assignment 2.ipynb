```python
import pandas as pd

df = pd.read_csv('cricket_data.csv')

df.head(2)
```

```python
df.info()        # See columns and data types
```

```python
df.describe()
```

```python
df['Highest_Score'] = df['Highest_Score'].str.replace('*', '', regex=False)
df['Runs_Scored'] = pd.to_numeric(df['Runs_Scored'], errors='coerce')
df['Batting_Average'] = pd.to_numeric(df['Batting_Average'], errors='coerce')

```

```python
# Who are the top 5 run scorer batsman?
top_scorers = df.groupby('Player_Name')['Runs_Scored'].sum().sort_values(ascending=False).head(5)
print("Top 5 Run Scorers:\n", top_scorers)
```

```python
# Who have Best Batting Averages (min 5 innings)?
df_filtered = df[df['Matches_Batted'] >= 5]
top_averages = df_filtered.groupby('Player_Name')['Batting_Average'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 Batting Averages:\n", top_averages)
```

```python
# Top 5 bowlers who took most Wickets
df['Wickets_Taken'] = pd.to_numeric(df['Wickets_Taken'], errors='coerce')
top_wickets = df.groupby('Player_Name')['Wickets_Taken'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Wicket Takers:\n", top_wickets)
```

```python
# What are total number of matches played by Ruturaj Gaikwad
# Converting Matches_Batted to numeric
df['Matches_Batted'] = pd.to_numeric(df['Matches_Batted'], errors='coerce')
# Filtering by player name and sum matches
ruturaj_matches = df[df['Player_Name'] == 'Ruturaj Gaikwad']['Matches_Batted'].sum()
print("Total matches played by Ruturaj Gaikwad:", int(ruturaj_matches))
```

```python
# How many players have at least 1 century?

df['Centuries'] = pd.to_numeric(df['Centuries'], errors='coerce')

total_centuries = df.groupby('Player_Name')['Centuries'].sum()

century_scorers = total_centuries[total_centuries >= 1].sort_values(ascending=False)

print("Players with at least 1 century:")
print(century_scorers)
```

```python
# Which player batted in the most matches?

df['Matches_Batted'] = pd.to_numeric(df['Matches_Batted'], errors='coerce')
batting_counts = df.groupby('Player_Name')['Matches_Batted'].sum()
# Finding the player with the most matches batted
most_matches_player = batting_counts.sort_values(ascending=False).head(1)
print("Player who batted in the most matches:")
print(most_matches_player)
```

```python
# Which players had hit maximum sixes and fours?


df['Sixes'] = pd.to_numeric(df['Sixes'], errors='coerce')
df['Fours'] = pd.to_numeric(df['Fours'], errors='coerce')

# Group by player and sum sixes and fours
sixes_by_player = df.groupby('Player_Name')['Sixes'].sum()
fours_by_player = df.groupby('Player_Name')['Fours'].sum()

# Find player with highest sixes
top_sixes_player = sixes_by_player.sort_values(ascending=False).head(1)

# Find player with highest fours
top_fours_player = fours_by_player.sort_values(ascending=False).head(1)

# Display results
print("Player with most sixes:")
print(top_sixes_player)

print("\nPlayer with most fours:")
print(top_fours_player)
```

```python
# Which player bowled the most balls?
# Convert Balls_Bowled to numeric
df['Balls_Bowled'] = pd.to_numeric(df['Balls_Bowled'], errors='coerce')

# Group by player and sum Balls_Bowled
balls_bowled_by_player = df.groupby('Player_Name')['Balls_Bowled'].sum()

# Find player with the highest Balls_Bowled
most_balls_bowled_player = balls_bowled_by_player.sort_values(ascending=False).head(1)

# Display the result
print("Player who bowled the most balls:")
print(most_balls_bowled_player)
```

```python
# Which player took maximum catches?

df['Catches_Taken'] = pd.to_numeric(df['Catches_Taken'], errors='coerce')

catches_by_player = df.groupby('Player_Name')['Catches_Taken'].sum()
most_catches_player = catches_by_player.sort_values(ascending=False).head(1)

print("Player with the most catches:")
print(most_catches_player)
```

```python
# Player with highest batting strike rate (min 10 matches)
df['Matches_Batted'] = pd.to_numeric(df['Matches_Batted'], errors='coerce')
df['Batting_Strike_Rate'] = pd.to_numeric(df['Batting_Strike_Rate'], errors='coerce')

filtered_df = df[df['Matches_Batted'] >= 10]

avg_strike_rate = filtered_df.groupby('Player_Name')['Batting_Strike_Rate'].mean()
top_strike_player = avg_strike_rate.sort_values(ascending=False).head(1)

print("Player with highest batting strike rate (min 10 matches):")
print(top_strike_player)
```
